## Κύριος Στόχος (Διάγραμμα LMM)

Το παρόν script **δεν εκτελεί ανάλυση δεδομένων**, αλλά έχει ως αποκλειστικό σκοπό τη δημιουργία μιας **οπτικοποίησης (διαγράμματος)** που να αποτυπώνει συνοπτικά τη λογική της μεθοδολογίας.

Ο στόχος είναι η παραγωγή ενός σαφούς **σχηματικού διαγράμματος** που απεικονίζει τη σχέση μεταξύ:

- της **κεντρικής στατιστικής μεθοδολογίας** που χρησιμοποιήθηκε – τα **Γραμμικά Μεικτά Μοντέλα (LMM)**,  
- και των διαφορετικών **φωνητικών tasks (εργασιών)** στα οποία αυτή η μεθοδολογία εφαρμόστηκε.

## Τεχνολογία

Για την υλοποίηση του διαγράμματος χρησιμοποιούνται:

- Η βιβλιοθήκη **`networkx`** για τη δημιουργία και διαχείριση της δομής του γράφου (κόμβοι και ακμές),
- Η βιβλιοθήκη **`matplotlib`** για τη σχεδίαση και την τελική απόδοση της εικόνας.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Δημιουργία γράφου
G_lmm = nx.DiGraph()

# Κόμβοι
model = "Γραμμικά Μεικτά Μοντέλα (LMM)"
tasks = [
    "Task 1:\nΣυνεχής εκφώνηση\nφωνηέντων/συμφώνων",
    "Task 2:\nΕκφώνηση συγκεκριμένων\nλέξεων/φράσεων",
    "Task 4:\nΦωνητικές ασκήσεις\nσυντονισμού αρθρώσεων",
    "Task 5:\nΕκφώνηση λέξεων με\nσυγκεκριμένα φωνήματα"
]

# Προσθήκη κόμβων και ακμών
G_lmm.add_node(model, layer=0)
for task in tasks:
    G_lmm.add_node(task, layer=1)
    G_lmm.add_edge(model, task)

# Σχεδίαση του διαγράμματος με μεγαλύτερη γραμματοσειρά
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G_lmm, seed=42, k=1.2)  # λίγο πιο αραιή διάταξη

nx.draw(
    G_lmm, pos,
    with_labels=True,
    node_color="lightblue",
    edge_color="black",
    node_size=5000,  # πιο μεγάλοι κόμβοι
    font_size=12,    # πιο μεγάλα γράμματα
    font_weight="bold",
    arrows=True,
    width=1.5
)

# Αν θέλεις και τίτλο στο γράφημα, μπορείς να τον ξεκλειδώσεις:
# plt.title("Γραμμικά Μεικτά Μοντέλα (LMM) και Tasks", fontsize=16, fontweight='bold')

plt.savefig("LMM_Diagram.png", dpi=300, bbox_inches='tight')

plt.show()
```


## Κύριος Στόχος (Διάγραμμα Ροής Εργασιών)

Το παρόν script αποσκοπεί αποκλειστικά στη δημιουργία μιας **οπτικοποίησης**:  
ενός **διαγράμματος ροής (flowchart)** που συνοψίζει ολόκληρη τη **μεθοδολογική διαδικασία μηχανικής μάθησης** που ακολουθήθηκε στη διπλωματική εργασία.

Ο στόχος είναι να παραχθεί μια **σαφής και περιεκτική οπτική αναπαράσταση** που να απεικονίζει τα βασικά στάδια της ανάλυσης.  
Αυτά τα στάδια περιλαμβάνουν:

- Τη **συλλογή δεδομένων**
- Την **προεπεξεργασία**
- Την **επιλογή χαρακτηριστικών**
- Την **εκπαίδευση μοντέλων** με τη μέθοδο **Leave-One-Subject-Out (LOSO)**
- Την **αξιολόγηση της απόδοσης**
- Την **παρουσίαση των αποτελεσμάτων**

## Τεχνολογία

Για την υλοποίηση του διαγράμματος χρησιμοποιείται η βιβλιοθήκη **`graphviz`**,  
η οποία επιτρέπει τον προγραμματιστικό ορισμό της δομής και της εμφάνισης του γράφου.  
Αυτή η προσέγγιση εξασφαλίζει **εύκολη τροποποίηση και αναπαραγωγή** του διαγράμματος.

```python
# --- 1. Εισαγωγή Βιβλιοθηκών ---
from graphviz import Digraph
from IPython.display import Image, display

# --- 2. Ορισμός της Συνάρτησης Δημιουργίας του Διαγράμματος ---
def generate_clean_flowchart():
    dot = Digraph(format='png')
    dot.attr(rankdir='TB', size='10')

    # --- 3. Ορισμός Κόμβων (Nodes) ---
    dot.node('A', 'Συλλογή & Οργάνωση Δεδομένων', shape='parallelogram', style='filled', fillcolor='lightblue')
    dot.node('B', 'Προεπεξεργασία Δεδομένων', shape='parallelogram', style='filled', fillcolor='lightgray')
    dot.node('C', 'Επιλογή Χαρακτηριστικών', shape='parallelogram', style='filled', fillcolor='lightyellow')
    dot.node('D', 'Εκπαίδευση Μοντέλων (LOSO)', shape='parallelogram', style='filled', fillcolor='lightgreen')
    dot.node('E', 'Αξιολόγηση Απόδοσης', shape='parallelogram', style='filled', fillcolor='orange')
    dot.node('F', 'Αποθήκευση & Παρουσίαση Αποτελεσμάτων', shape='parallelogram', style='filled', fillcolor='lightcoral')

    # Δημιουργία ενός υπο-γράφου (subgraph) για την ομαδοποίηση των μοντέλων μηχανικής μάθησης.
    # Το `rank='same'` τους τοποθετεί στο ίδιο οριζόντιο επίπεδο.
    with dot.subgraph() as s:
        s.attr(rank='same')
        s.node('D1', 'LR', shape='box', style='filled', fillcolor='lightgreen')
        s.node('D2', 'RF', shape='box', style='filled', fillcolor='lightgreen')
        s.node('D3', 'SVM', shape='box', style='filled', fillcolor='lightgreen')
        s.node('D4', 'KNN', shape='box', style='filled', fillcolor='lightgreen')
        s.node('D5', 'XGBoost', shape='box', style='filled', fillcolor='lightgreen')

    # --- 4. Ορισμός Συνδέσεων (Edges) ---
    dot.edge('A', 'B', label='Καθαρισμός & Κανονικοποίηση')
    dot.edge('B', 'C', label='Φιλτράρισμα με p-value < 0.05')
    dot.edge('C', 'D', label='Χρήση επιλεγμένων χαρακτηριστικών')

    # Βρόχος για τη σύνδεση του γενικού σταδίου 'Εκπαίδευση' με κάθε συγκεκριμένο μοντέλο,
    # και κάθε μοντέλου με το στάδιο 'Αξιολόγηση'.
    for model in ['D1', 'D2', 'D3', 'D4', 'D5']:
        dot.edge('D', model)
        dot.edge(model, 'E')

    dot.edge('E', 'F', label='Σύγκριση & Οπτικοποίηση αποτελεσμάτων')

    # --- 5. Εμφάνιση του Διαγράμματος ---
    display(Image(dot.pipe(format='png')))

# --- 6. Εκτέλεση της Συνάρτησης ---
generate_clean_flowchart()
```

## Κύριος Στόχος (Διάγραμμα για eGeMAPS)

Ο στόχος είναι να παραχθεί ένα διάγραμμα τύπου **mind map** που να δείχνει ότι το  
**eGeMAPS** (*extended Geneva Minimalistic Acoustic Parameter Set*) είναι ένα  
σετ χαρακτηριστικών **υψηλού επιπέδου**, το οποίο προκύπτει από τη στατιστική  
ανάλυση (π.χ. μέσος όρος, τυπική απόκλιση) πολλών διαφορετικών **κατηγοριών  
ακουστικών χαρακτηριστικών χαμηλού επιπέδου** (*Low-Level Descriptors - LLDs*).

## Τεχνολογία

Το διάγραμμα κατασκευάζεται εξ ολοκλήρου με τη χρήση της βιβλιοθήκης `matplotlib`.  
Η **κυκλική διάταξη** των περιφερειακών κόμβων υπολογίζεται μαθηματικά για την  
επίτευξη συμμετρίας και αισθητικής αρτιότητας. Κάθε στοιχείο (κύκλοι, γραμμές,  
πλαίσια κειμένου) τοποθετείται με ακρίβεια στον καμβά.


```python
# --- 1. Εισαγωγή Βιβλιοθηκών ---
import matplotlib.pyplot as plt
import numpy as np

# --- 2. Ορισμός Περιεχομένου και Διάταξης ---
# Ορισμός των ετικετών (κειμένων) για τους περιφερειακούς κόμβους του διαγράμματος.
labels = [
    "Φωνητικά χαρακτηριστικά\n→ Jitter, Shimmer, HNR,\nAlpha Ratio",
    "Φασματικά χαρακτηριστικά\n→ MFCCs, spectral slope,\nspectral flux",
    "Ενεργειακά χαρακτηριστικά\n→ Loudness,\nAuditory Spectrum",
    "Ρυθμικά / χρονικά\n→ Διάρκεια παύσεων,\nλόγος ομιλίας / σιωπής",
    "Στατιστικά μέτρα\n→ Μέσος όρος,\nΤ.Α., percentiles"
]

# Μαθηματικός υπολογισμός για την κυκλική διάταξη των κόμβων.
num_labels = len(labels)
radius = 3.5
angles = np.linspace(0, 2 * np.pi, num_labels, endpoint=False)
positions = [(radius * np.cos(a), radius * np.sin(a)) for a in angles]

# --- 3. Δημιουργία και Σχεδίαση του Διαγράμματος ---
plt.figure(figsize=(9, 9))  # Ελαφρώς μεγαλύτερο για να έχεις καλύτερη ανάλυση
ax = plt.gca()
ax.set_facecolor('white')

# --- Σχεδίαση του κεντρικού κόμβου ---
center_x, center_y = 0, 0
circle_radius = 0.8
circle = plt.Circle((center_x, center_y), circle_radius, color='#BFE8C6', ec='black', zorder=2)
ax.add_patch(circle)
plt.text(center_x, center_y, 'eGeMAPS', ha='center', va='center', fontsize=20, fontweight='bold', zorder=3)

# --- Σχεδίαση των περιφερειακών κόμβων και των γραμμών σύνδεσης ---
for (x, y), text, angle in zip(positions, labels, angles):
    # Υπολογισμός νέου σημείου εκκίνησης γραμμής στην περιφέρεια του κύκλου
    line_start_x = center_x + circle_radius * np.cos(angle)
    line_start_y = center_y + circle_radius * np.sin(angle)
    plt.plot([line_start_x, x], [line_start_y, y], color='black', lw=1)
    plt.text(
        x, y, text,
        ha='center', va='center', fontsize=14, fontweight='bold',
        bbox=dict(boxstyle="round,pad=0.6", facecolor='#D8F2DF', edgecolor='black')
    )

# --- 4. Τελικές Ρυθμίσεις και Αποθήκευση ---
plt.axis('off')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.tight_layout()
plt.savefig("eGeMAPS_Clean_BiggerFont.png", dpi=300, bbox_inches='tight')
plt.show()
```
